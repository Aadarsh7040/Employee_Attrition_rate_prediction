```python
from scipy.stats import chi2_contingency
stat, p, dof, expected = chi2_contingency(A)
p
```

```python
stat, p, dof, expected
```

```python
# interpret p-value
alpha = 0.05
print("p value is " + str(p))
if p <= alpha:
    print('reject H0)')
else:
    print('Accept H0')
```

# 8. Form Classification Model using Y=attrition and choose the best model

```python
Y=df[['Attrition']]
Y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=31,stratify=Y)
xtrain
```

## Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
log_model=LogisticRegression()
log_model.fit(xtrain,ytrain)
```

```python
#TRAINING
ypredtrain=log_model.predict(xtrain)
accuracy=accuracy_score(ytrain,ypredtrain)
print('Training Accuracy :',accuracy*100)
score=cross_val_score(log_model,X,Y,cv=5)
print('cross validation value',np.mean(score))
confusionmatrix = confusion_matrix(ytrain,ypredtrain)
print('Confusion matrix: \n',confusionmatrix)
Classification_report = classification_report(ytrain,ypredtrain)
print('Classification report:\n',Classification_report)

# TESTING
ypredtest=log_model.predict(xtest)
accuracy=accuracy_score(ytest,ypredtest)
print('Testing Accuracy :',accuracy*100)
score=cross_val_score(log_model,X,Y,cv=5)
print('cross validation value',np.mean(score))
confusionmatrix = confusion_matrix(ytest,ypredtest)
print('Confusion matrix: \n',confusionmatrix)
Classification_report = classification_report(ytest,ypredtest)
print('Classification report:\n',Classification_report)
```

### Within cluster sum of squares (wccs)

```python
WCCS = kmeans_model.inertia_  # k = 2
WCCS
```

## Elbow Method

```python
wcss_list = []
for k in range(1,10):
    kmeans_model = KMeans(n_clusters=k)
    kmeans_model.fit(x)
    wcss = kmeans_model.inertia_
    wcss_list.append(wcss)
    print('K =',k,  'WCSS = ',wcss)
    
```

```python
wcss_list
```

```python
k = range(1,10)
plt.plot(k,wcss_list)
plt.xlabel('Number of cluster(k)')
plt.ylabel('WCSS')
plt.title('Elbow method graph')
```
